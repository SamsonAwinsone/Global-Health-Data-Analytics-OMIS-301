```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset from an Excel file
df = pd.read_excel('C:/Users/HP/Documents/Class_work_data.xlsx', sheet_name='Sheet1')

# Task 1: Exploratory Data Analysis (EDA)
# Summary statistics for numerical and categorical columns
numerical_summary = df.describe()
categorical_summary = df.describe(include=['object'])

plt.figure(figsize=(12, 6))

# Prevalence Rate distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Prevalence Rate (%)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Prevalence Rate (%)')
plt.xlabel('Prevalence Rate (%)')
plt.ylabel('Frequency')

# Mortality Rate distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Mortality Rate (%)'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Mortality Rate (%)')
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Bar chart comparing Healthcare Access (%) across the countries
plt.figure(figsize=(14, 6))
healthcare_access = df.groupby('Country')['Healthcare Access (%)'].mean().sort_values()
healthcare_access.plot(kind='barh', color='lightgreen')
plt.title('Healthcare Access (%) by Country')
plt.xlabel('Healthcare Access (%)')
plt.ylabel('Country')
plt.show()

# Display summary statistics
numerical_summary, categorical_summary
```

```python
# Task 2: Insights on Disease Prevalence and Healthcare Investment

# Top 3 diseases with the highest prevalence rate
top_diseases = df.groupby('Disease Name')['Prevalence Rate (%)'].mean().sort_values(ascending=False).head(3)

# Countries requiring more investment based on Doctors per 1000 and Hospital Beds per 1000
low_resources = df[df['Doctors per 1000'] < 2].groupby('Country')[['Doctors per 1000', 'Hospital Beds per 1000']].mean()

# Visualizing the findings
plt.figure(figsize=(10, 6))
top_diseases.plot(kind='bar', color='orange')
plt.title('Top 3 Diseases with Highest Prevalence Rate (%)')
plt.xlabel('Disease Name')
plt.ylabel('Average Prevalence Rate (%)')
plt.xticks(rotation=45)
plt.show()

low_resources
```

```python
# Task 3: Correlation Analysis

# Selecting relevant columns for correlation analysis
correlation_data = df[['Education Index', 'Per Capita Income (USD)', 'Prevalence Rate (%)']]

#Correlation coefficients
correlation_matrix = correlation_data.corr()

# Scatter plots for visualizing relationships
plt.figure(figsize=(15, 5))

# Education Index vs. Prevalence Rate (%)
plt.subplot(1, 3, 1)
sns.scatterplot(x='Education Index', y='Prevalence Rate (%)', data=df, color='blue')
plt.title('Education Index vs. Prevalence Rate (%)')

# Per Capita Income (USD) vs. Prevalence Rate (%)
plt.subplot(1, 3, 2)
sns.scatterplot(x='Per Capita Income (USD)', y='Prevalence Rate (%)', data=df, color='green')
plt.title('Per Capita Income (USD) vs. Prevalence Rate (%)')

# Education Index vs. Per Capita Income (USD)
plt.subplot(1, 3, 3)
sns.scatterplot(x='Education Index', y='Per Capita Income (USD)', data=df, color='purple')
plt.title('Education Index vs. Per Capita Income (USD)')

plt.tight_layout()
plt.show()

# Task 4: Targeted Interventions
# Identify regions with the highest Disease Burden (DALYs)
high_disease_burden = df.groupby('Country')['DALYs'].sum().sort_values(ascending=False).head(5)

# Visualize Disease Burden for top countries
plt.figure(figsize=(10, 6))
high_disease_burden.plot(kind='bar', color='red')
plt.title('Top 5 Countries with Highest Disease Burden (DALYs)')
plt.xlabel('Country')
plt.ylabel('Total DALYs')
plt.xticks(rotation=45)
plt.show()

# Display the correlation matrix and high disease burden data
correlation_matrix, high_disease_burden
```
